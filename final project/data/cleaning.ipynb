```python
import pandas as pd
import numpy as np

# 1. 重新讀取原始資料
df = pd.read_csv('YRBS_2007.csv')

# 2. 挑選需要的欄位並先複製一份
cleaned_df = df[['InWhatGradeAreYou', 'AerobicExercise']].copy()

# 3. 重新命名欄位
cleaned_df.columns = ['Grade', 'Aerobic_Exercise']

# 4. 【核心修正】強制將資料轉為數字，無法轉換的（例如空白、文字錯誤）會變成 NaN
cleaned_df['Grade'] = pd.to_numeric(cleaned_df['Grade'], errors='coerce')
cleaned_df['Aerobic_Exercise'] = pd.to_numeric(cleaned_df['Aerobic_Exercise'], errors='coerce')

# 5. 剔除任何含有 NaN (缺失值) 的整列資料
cleaned_df = cleaned_df.dropna()

# 6. 建立年級的對應字典（同時相容問卷代碼 1~4 與實際年級 9~12）
grade_mapping = {
    1: '9th Grade', 2: '10th Grade', 3: '11th Grade', 4: '12th Grade',
    9: '9th Grade', 10: '10th Grade', 11: '11th Grade', 12: '12th Grade'
}

# 進行對應轉換
cleaned_df['Grade'] = cleaned_df['Grade'].map(grade_mapping)

# 7. 再剔除一次沒有成功對應到年級的無效資料，並確保運動天數是整數
cleaned_df = cleaned_df.dropna()
cleaned_df['Aerobic_Exercise'] = cleaned_df['Aerobic_Exercise'].astype(int)

# 8. 儲存成乾淨的 CSV 檔
cleaned_df.to_csv('cleaned_yrbs.csv', index=False)

# 9. 印出結果檢查
print(f"原始資料筆數: {df.shape[0]} 筆")
print(f"清洗後有效資料: {cleaned_df.shape[0]} 筆")
print("\n--- 各年級清洗後的有效樣本數 ---")
print(cleaned_df['Grade'].value_counts())
```
